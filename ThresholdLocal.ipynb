```python
#import library
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def TheresholdLocalFindT(img_gray, delta_T):
  T = np.mean(img_gray)
  T_pve = 0
  while abs(T - T_pve) >=delta_T:
    G1 = img_gray[img_gray > T]
    G2 = img_gray[img_gray <= T]
    m1 = np.mean(G1)
    m2 = np.mean(G2)
    T_pve = T
    T = (m1 + m2)/2
  return T

def  Thereshold(img, delta_T):
  T = TheresholdLocalFindT(img, delta_T)
  img[img >= T] = 255
  img[img < T] = 0
  return img
```

```python
file = "/content/A.jpg"
delta_T = 1

img = cv.imread(file)
print("Ảnh Gốc")
cv2_imshow(img)
thereshold = Thereshold(img, delta_T)
print("")
print("Ảnh sau khi cắt ngưỡng")
cv2_imshow(thereshold)
cv.waitKey()
cv.destroyAllWindows()
```
